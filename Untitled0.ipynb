## Prepare Environment and Define Sentences

### Subtask:
Install necessary libraries (sentence-transformers), import them, and define 10 sentences across Cricket, Cooking, and Cybersecurity topics as specified in the prompt. I will ensure at least 3 sentences per topic.


**Reasoning**:
First, I need to install the `sentence-transformers` library, which is a prerequisite for the subtask. This step needs to be executed before any imports or sentence definitions.

**Reasoning**:
Now that the library is installed, I need to import the `SentenceTransformer` class and define the `sentences` list as specified in the subtask, ensuring at least 3 sentences per topic (Cricket, Cooking, Cybersecurity).



```python
from sentence_transformers import SentenceTransformer

sentences = [
    # Cricket
    'The batsman hit a magnificent century to lead his team to victory.',
    'The fast bowler delivered a perfect yorker at the death overs.',
    'Cricket fans around the world are excited for the upcoming World Cup.',
    'The umpire raised his finger, signaling an out.',

    # Cooking
    'Always sauté the onions until they turn golden brown for the best flavor.',
    'Baking bread requires a precise balance of flour, water, and yeast.',
    'Slow-cooking the meat makes it incredibly tender and juicy.',
    'Add a pinch of salt and black pepper to season the dish properly.',

    # Cybersecurity
    'Cybersecurity is crucial for protecting sensitive data from breaches.',
    'Multi-factor authentication adds an extra layer of security to online accounts.',
    'Phishing attacks are a common threat, often disguised as legitimate emails.',
    'Regular software updates are essential to patch security vulnerabilities.'
]

print(f"Defined {len(sentences)} sentences.")
print("First 3 sentences:")
for i in range(3):
    print(f"- {sentences[i]}")
```

## Generate Embeddings

### Subtask:
Initialize the `sentence-transformers` model `all-MiniLM-L6-v2` and generate embeddings for all the defined sentences. Also, generate an embedding for the query sentence 'The bowler took three wickets in one over'.


**Reasoning**:
I need to initialize the SentenceTransformer model, generate embeddings for both the predefined sentences and the query sentence, and then print the shapes of these embeddings to confirm they were generated correctly. All these actions can be grouped into a single code block.



```python
model = SentenceTransformer('all-MiniLM-L6-v2')

sentence_embeddings = model.encode(sentences)

query_sentence = 'The bowler took three wickets in one over'
query_embedding = model.encode(query_sentence)

print(f"Shape of sentence_embeddings: {sentence_embeddings.shape}")
print(f"Shape of query_embedding: {query_embedding.shape}")
```

**Reasoning**:
The previous step successfully generated embeddings. Now, I need to compute the cosine similarity matrix for these sentence embeddings and visualize it as a heatmap as per the task requirements.



```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sentence_transformers import util

# Compute cosine similarity between all sentence embeddings
cosine_scores = util.cos_sim(sentence_embeddings, sentence_embeddings)

# Convert to numpy array for seaborn heatmap
cosine_scores_np = cosine_scores.numpy()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(cosine_scores_np, annot=True, cmap='viridis', fmt=".2f",
            xticklabels=[f'S{i+1}' for i in range(len(sentences))],
            yticklabels=[f'S{i+1}' for i in range(len(sentences))])
plt.title('Cosine Similarity Matrix of Sentences')
plt.xlabel('Sentence Index')
plt.ylabel('Sentence Index')
plt.show()

print("Cosine similarity matrix computed and visualized.")
```

**Reasoning**:
I need to calculate the cosine similarity between the query sentence embedding and all original sentence embeddings, then identify and print the top 2 most similar sentences along with their scores.



```python
from sentence_transformers import util

# Compute cosine similarity between the query embedding and all sentence embeddings
query_scores = util.cos_sim(query_embedding, sentence_embeddings)[0]

# Combine scores with original sentences and sort to find top 2
sentence_scores = []
for i, score in enumerate(query_scores):
    sentence_scores.append({'sentence': sentences[i], 'score': score.item()})

sentence_scores = sorted(sentence_scores, key=lambda x: x['score'], reverse=True)

print(f"Query sentence: '{query_sentence}'\n")
print("Top 2 most similar sentences:")
for i, hit in enumerate(sentence_scores[0:2]):
    print(f"  {i+1}: {hit['sentence']} (Score: {hit['score']:.4f})")
```
